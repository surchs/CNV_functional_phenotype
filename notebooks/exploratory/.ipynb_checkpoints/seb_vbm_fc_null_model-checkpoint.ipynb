```python
def octvec2mat(vec, mask):
    """
    Takes a vector in Fortran style (Octave/Matlab)
    and maps it back into a mask correctly in 
    python
    """
    vec_mat = mask.flatten(order='F').astype(bool)
    tmp = np.zeros_like(vec_mat, dtype=float)
    tmp[vec_mat] = vec
    vol = np.reshape(tmp, mask.shape, order='F')
    return vol
```

```python
def vec2mat(vec, mask):
    tmp_mat = octvec2mat(vec, mask)
    tmp_mat += tmp_mat.T
    tmp_mat[np.eye(mask.shape[0]).astype(bool)] = tmp_mat[np.eye(mask.shape[0]).astype(bool)]/2
    return tmp_mat
```

## Test correlation of random vectors
Use two random vectors and correlate them but vary the amount of overlap

```python
c = np.zeros(100)
for i in range(100):
    a = np.zeros(100000)
    b = np.zeros(100000)
    a[:100] = np.random.random(100)
    b[i:i+100] = np.random.random(100)
    c[i] = sst.pearsonr(a,b)[0]
```

```python
plt.plot((100-np.arange(100)), c)
plt.xlabel('Percent Overlap')
plt.ylabel('Pearson Correlation')
```

Here we get close to perfect correlation between two strictly positive random vectors that have a long end of zero values. Very similar to what I saw with the brain

```python
n = np.zeros(100)
for i in range(100):
    a = np.zeros(100000)
    b = np.zeros(100000)
    a[:100] = np.random.random(100)*2-1
    b[i:i+100] = np.random.random(100)*2-1
    n[i] = sst.pearsonr(a,b)[0]

plt.plot((100-np.arange(100)), n);
plt.xlabel('Percent Overlap');
plt.ylabel('Pearson Correlation');
```

This effect disappears when I allow the vectors to have negative values

```python
a = np.zeros(100000)
b = np.zeros(100000)
a[:100] = np.random.random(100)
b[50:150] = np.random.random(100)
sst.pearsonr(a,b)
```

```python
plt.plot((a-np.mean(a))[:200]);
plt.title('Residuals of A');
```

```python
plt.plot((b-np.mean(b))[:200]);
plt.title('Residuals of B');
```

```python
plt.plot(((a-np.mean(a))*(b-np.mean(b)))[:200]);
plt.title('Product of residuals of A and B for overlap');
```

```python
plt.plot(((a-np.mean(a))*(b-np.mean(b)))[200:400]);
plt.title('Product of residuals of A and B without overlap');
```

So what is happening here? Both A and B are strictly positive and have a huge number of zero values. These zero values pull the mean towards 0:

```python
print(np.mean(a), np.mean(b))
```

Pretty close to 0 actually. By contrast, the random values range between 0 and 1, most of them will be larger than the small mean. So most of the residuals are positive. Therefore, the covariance of the overlap is also positive (in both maps, residuals have the same sign). But the same is true for all the zeros. They are all (somewhat) smaller than the mean - in both maps. So again, the covariance is positive (residuals in both maps are negative...). So why does the amount of overlap even matter given how many zeros there are with positive covariance? The covariance of the overlap is orders of magnitude larger than that of the zeros, and it's relative contribution is larger if the zone of overlap is larger. Since the whole thing gets standardized by the standard deviation, scaling the random values has little effect. Setting the zeros to an arbitrarily negativ value does help however since it makes the residuals of the overlap region much larger up to the point where their randomness doesn't matter anymore and it essentially becomes a binary vector.

```python
a = np.zeros(100000)
b = np.zeros(100000)
a[:100] = np.random.random(100)
a[100:] = -100000
b[:100] = np.random.random(100)
b[100:] = -100000
sst.pearsonr(a,b)
```

```python
a = np.zeros(100000)
b = np.zeros(100000)
a[:100] = 1
b[:100] = 1
sst.pearsonr(a,b)
```

Basically the same story.

So what does this mean for my analysis? 
1. Including zeros that can never be non-zero inflates the correlation estimate if the maps themselves can't be negative
2. This is because the zeros pull the average down and make more residuals positive, increasing covariance
3. Excluding the non-brain voxels brings the analysis closer to what I intended and measures the distribution of effect sizes

__BUT__: this becomes hard to interpret since we change the number of significant regions on the FC side. If all regions have non-zero effects on the FC side, then the highest values have to be where the VBM effects are or else we will even get a negative correlation. On the other hand, if almost no regions are non-zero, even small effects in the right place will have positive correlation. Couldn't we just use the DICE here? Not sure. This measures something else again.
