## Severity
Procedure:
- get the TOTAL score, the age and the module
- if the age and module are a match, look up the css and keep it
- if the subject is too old for the module, see if they got module 4
    - if yes, just give them the maximal acceptable age in the module and use that score
    - if it's not module 4, see if the subject is within 2 years of the age cap
        - if yes, give them the maximal acceptable age of their module
        - if not, don't give them anything. kick them out

```python
severity = list()
for rid, row in pheno.iterrows():
    # See if we have a TOTAL score and module information
    if not np.isnan(row['ADOS_TOTAL']) and not np.isnan(row['ADOS_MODULE']):
        # Look up the score
        age = np.round(row['AGE_AT_SCAN'])
        module = row['ADOS_MODULE']
        total = row['ADOS_TOTAL']
        # Check if the age is appropriate for the module
        ind_module = (sev_lut['module']==module).values
        age_max = np.max(sev_lut[ind_module]['age_max'].astype(float).values)
        age_min = np.min(sev_lut[ind_module]['age_min'].astype(float).values)
        if age>age_max:
            # This person got the wrong module
            # See if it is module 4
            if module==4:
                # Fine, just take whatever is the max age, they are old anyway
                print('{} >>>> {} in module {}'.format(age, age_max, module))
                age = age_max
            # Otherwise see if it is within 2 years of the max
            elif age-age_max < 3:
                print('{} >>>> {} in module {}'.format(age, age_max, module))
                age = age_max
            else:
                # Do nothing
                pass
        elif age<age_min:
            print('!!!!!oh oh!!!!')
                        
        # Find the correct element
        
        ind = ((sev_lut['module']==module).values & 
               (sev_lut['age_min'].astype(float)<= age).values & 
               (sev_lut['age_max'].astype(float)>= age).values &
               (sev_lut['val_min'].astype(float)<= total).values &
               (sev_lut['val_max'].astype(float)>= total).values)
        val = sev_lut[ind]['css'].values
        # See if we actually found anything
        if not val:
            print('    Out with you!')
            print('   ', age, age_max, module, total)
            severity.append(None)
        else:
            css = val[0]
            severity.append(css)
    else:
        severity.append(None)
```

To summarize: nobody was too young for their model. Several individuals were too old for their model but many were within 2 years of the age cap and were re-assigned. In total, 4 subjects were outside of my acceptance limits and got kicked out.

```python
# Put the estimated ADOS_GOTHAM scores in the table
model['ADOS_GOTHAM_ESTIMATED_sev'] = severity
```
