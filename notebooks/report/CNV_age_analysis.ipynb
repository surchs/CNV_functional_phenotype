```python
def norm(array):
    return (array-np.mean(array))/np.std(array)
```

```python
def octvec2mat(vec, mask):
    """
    Takes a vector in Fortran style (Octave/Matlab)
    and maps it back into a mask correctly in 
    python
    """
    vec_mat = mask.flatten(order='F').astype(bool)
    tmp = np.zeros_like(vec_mat, dtype=float)
    tmp[vec_mat] = vec
    vol = np.reshape(tmp, mask.shape, order='F')
    return vol
```

```python
def vec2mat(vec, mask):
    tmp_mat = octvec2mat(vec, mask)
    tmp_mat += tmp_mat.T
    tmp_mat[np.eye(mask.shape[0]).astype(bool)] = tmp_mat[np.eye(mask.shape[0]).astype(bool)]/2
    return tmp_mat
```

```python
def model_fit(data, model):
    p = list()
    b = list()
    for i in range(2080):
        mod = sm.OLS(data[:, i], model.values)
        res = mod.fit()
        b.append(res.params[-1])
        p.append(res.pvalues[-1])
    return p, b
```

## Age-match without regressing
how do the CNV findings change when we only match by age but don't regress age. So here we separate the question of power from the question of age-signal. 

```python
# age matched by not regressed
model16_young = pat.dmatrix('sex_dummy+FD_scrubbed_both_norm+Site_dummy+g1', data=pheno_16_del_con_young, return_type='dataframe')
(p16_young, b16_young) = model_fit(conn16_del_con[age_mask, :], model16_young)
```

The red line represents a theoretical perfect correspondence between full and age-matched sample while the black line represents the real relationship.

```python
print('The correlation between the effects in the whole sample and'
      ' the age matched sample is {:.2f} (p={:.2E})'.format(*sst.pearsonr(b16_ref, b16_young)))
```

```python
sig_young = len(np.where(stm(p16_young, alpha=0.05, method='fdr_bh')[0])[0])
overlap_young_all = len(np.intersect1d(np.where(stm(p16_young, alpha=0.05, method='fdr_bh')[0])[0],
                                        np.where(stm(p16_base, alpha=0.05, method='fdr_bh')[0])[0]))
```

```python
print('{} connections survive FDR in the age-matched sample. '
      'Of those, {} are also connections that survive FDR if we don\'t age-match.'.format(sig_young,
                                                                                                overlap_young_all))
```

```python
g = sbn.regplot(b16_ref, b16_young-b16_ref, marker='.', color='grey', line_kws={'color':'black'})
g.set_xlabel('FC effect in whole sample')
g.set_ylabel('change in FC effect in young sample')
g.set_title('FC contrast changes as function of effect in large sample');
```

```python
print('The correlation between the effects in the whole sample and'
      ' the change of effects in the age matched sample is {:.2f} (p={:.2E})'.format(*sst.pearsonr(b16_ref, b16_young-b16_ref)))
```

So here is what this tells us:
- there is a strong similarity of the CNV effects we find the in the whole and in the age matched sample (r=0.86, p=0)
- but the remaining differences are not random. Rather they push effects towards zero (regression towards the mean) in that they make negative effects less negative and positive effects less positive
- since we don't correct for age in either sample (whole or young) here, this is not due to variance explained by age directly. Rather it is due to the differences in the control groups. That is, in the whole sample we compare young DELs to young and old CONs whereas we compare them to only young CONs in the young group. 

Since the DELs are (almost) identical (only 2 DELs are removed by the age threshold), we can assume that there is some systematic difference between controls below 200 months (16.6 years) and above. We can test that quickly.

# Regressing age effect
So what happens when we actually regress age in the GLM?

## In the whole sample

```python
# Regressing age from the whole sample
model16_age_base = pat.dmatrix('sex_dummy+FD_scrubbed_both_norm+Site_dummy+age_months_norm+g1', data=pheno_16_del_con, return_type='dataframe')
(p16_age_base, b16_age_base) = model_fit(conn16_del_con, model16_age_base)
```

The black line represents the linear fit through the data, the red line represents a theoretical perfect relationship. Where the uncorrected CNV effect (red line) is negative, the corrected effect (black line) is less negative (more positive). Where the uncorrected effect is positive, the corrected effect is less positive (more negative). So by correcting for age in the whole sample, we push our effects toward zero.

```python
print('The correlation between the CNV effect in the whole sample without correction for age \nand'
      ' the CNV effect after correcting for age is {:.2f} (p={:.2E})'.format(*sst.pearsonr(b16_ref, b16_age_base)))
```

In terms of profile similarity with 22q, correcting for age in the whole sample doesn't change much. # Summary of age in 16p
1. Controlling for age, either by age-matching or regressing age changes our findings in a systematic way that pushes strong effects towards zero
2. This is probably in part due to differences in connectivity between the younger and older control subjects that weakly inverted from the CNV effect in the whole sample. The age differences in the controls are not significant however.
3. Within the age-matched group, 13 connections survive FDR if we don't regress age. 5 connections survive FDR if we regress age. In either case, these connections also survive FDR in the original analysis (whole sample, no age).
4. Regressing age in the age-matched sample does not change the effects in a systematic way.
5. All of these strategies result in a profile of connectivity alterations that is similar to the 22q profile.
6. Correcting for age in the whole sample gives a profile that is very similar to the uncorrected vector profile. Correcting for age in the young population results in a vector that is similar to 22q but differently than the uncorrected vector.

# Overall summary
1. Correcting for age in 16p, either by age matching, age regression, or both, pushes results towards zero systematically
2. The vector similarity with 22q remains largely unaffected by age matching or age regression but combining both, leads to a different spatial pattern of similarity with 22q (although still high)
3. Each age-correction method reduces the number of significant connections but the remaining connections are for the most part a subset of the connections we found originally
4. The effect of age matching and age regression seem to be at least in part due to a difference of FC in the older 16p controls. They tend to be less like 16p DEL than the younger 16 controls. These are non-significant differences however.
5. There are significant differences between the age-matched 16p and 22q controls.
6. Those are more pronounced if we don't account for the variance due to age (although they are already age matched, the matching is pretty simple).
7. If we ignore the significant differences (we probably shouldn't) and the dependencies this introduces for comparing with the 22q vector (we probably shouldn't either), we find that adding the 22q controls to the age-matched 16p DEL-CON contrast, gives us 6 more significant connections (11 total) and quite a different pattern of FC changes.
8. Despite all this, the similarity with the 22q vector is still very similar.

__Bottom line__: There are no great solutions unless we find a large control sample that doesn't show significant differences to our young 16p controls. Short of that, we might be best off by just regressing age on the whole dataset.
